```python
import pandas as pd
import numpy as np

# File path
carbon_emissions_path = "C:\\Users\\ADMIN\\Desktop\\Carbon-Emisisions\\Emissions Dataset\\Carbon Dioxide Emission Estimates.csv"
reductions_path = "C:\\Users\\ADMIN\\Desktop\\Carbon-Emisisions\\Emissions Dataset\\Reductions Dataset.csv"
    
# Read CSV file into DataFrame
carbon_df = pd.read_csv(carbon_emissions_path)
reductions_df = pd.read_csv(reductions_path)
```

```python
# Create copy of the original dataframe
carbon_df_original = carbon_df.copy()
reductions_df_original = reductions_df.copy()
```

```python
# Carbon emissions
carbon_df["CO2 emission estimates"] = carbon_df["CO2 emission estimates"].astype("category")
carbon_df["Years"] = carbon_df["Years"].astype("int64")
carbon_df["Series"] = carbon_df["Series"].astype("category")
carbon_df["Value"] = pd.to_numeric(carbon_df["Value"], errors="coerce")

# Quick check of dtypes
carbon_df.info()
```

```python
# Carbon reductions

reductions_df['Type'] = reductions_df['Type'].astype('category')
reductions_df['Reduced Carbon(Gt)'] = reductions_df['Reduced Carbon(Gt)'].astype('float')
reductions_df['How'] = reductions_df['How'].astype('category')

# Quick check of dtypes

reductions_df.info()
```

```python
# Rename columns to be more descriptive and proper
carbon_df.rename(columns={
    'CO2 emission estimates': 'Country',
    'Years': 'Year',
    'Series': 'Emission_Type',
    'Value': 'CO2_Value'
}, inplace=True)
```

```python
# Check for null values in all columns
def check_null(carbon_df):
    null_counts = carbon_df.isnull().sum()
    total_nulls = null_counts.sum()
    return null_counts, total_nulls

print(check_null(carbon_df))
```
